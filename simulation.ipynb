```python
import scipy.optimize as optimize
```

```python
alpha = 1/95000
def F(x):
    return alpha*(x**2-900**2)/2
```

```python
def f(params):
    b1, b2 = params
    return -((1000-b1)*F(b1) + (1000-b2)*(F(b2)-F(b1)))

initial_guess = [900+100/3, 900+100/3*2]
result = optimize.minimize(f, initial_guess)
```

```python
result.x
```

```python
import numpy as np

n=200000
n_goldfishes = 10000

goldfishes = np.sqrt(2*np.random.uniform(0,1,size=(n,n_goldfishes))/alpha+900**2)
bids = np.random.uniform(900,1000,size=(n,2))
```

```python
bids.sort(axis=1)
```

```python
payoffs = np.where(goldfishes<bids[:,:1],1000-bids[:,:1],np.where(goldfishes<bids[:,1:], 1000-bids[:,1:], 0))
payoffs.mean(axis=1).mean()
```

```python
bids[np.argsort(((bids-np.array([933.93, 967.25]))**2).sum(axis=1))][0:100]
```

```python
bids[payoffs.mean(axis=1).argmax()]
```
